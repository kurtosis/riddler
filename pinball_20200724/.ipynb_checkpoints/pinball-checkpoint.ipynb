```python
import math
import numpy as np
import pandas as pd
import plotnine as pn
```

```python
INITIAL_POSITION = (-2.0, 2.0)
CIRCLE_CENTER = (0, 2.0)
RADIUS = 1.0
```

```python
def get_circle_contact(x_wall_start, angle=None, slope=None):
    """Find the point at which a ball leaving wall will contact the circle.
    Note that this function assumes circle centered at (0,2) with radius 1."""
    if slope is None:
        angle_init = np.radians(angle)
        slope = 1/np.tan(angle_init)
        print(f'slope {slope}')
    a = slope**2 + 1
    b = 2*slope*x_wall_start - 4
    c = x_wall_start**2 + 3    
    try:
        y_circle = np.roots((a,b,c))
    except:
        return x_wall_start+2.1*slope, 2.1, x_wall_start+3*slope, 3
    y_circle = [y for y in y_circle if not isinstance(y, complex)]
    if len(y_circle)==0:
        return x_wall_start+2.1*slope, 2.1, x_wall_start+3*slope, 3
    else:
        y_circle = min(y_circle)
        x_circle = x_wall_start + slope*y_circle

#         print(f'circle {x_circle, y_circle}')
        radial_vector = (x_circle, y_circle-2) 
        radial_vector = radial_vector/np.linalg.norm(radial_vector)
        angle_radius_horizon = np.arccos(radial_vector[0])
        angle_triangle = 2*np.pi - angle_init - angle_radius_horizon
        angle_circle_rebound = angle_init - (2*np.pi - angle_triangle)
        print(f'angle_circle_rebound {np.degrees(angle_circle_rebound)}')
        delta_x = -y_circle/np.tan(angle_circle_rebound)
        x_wall_final = x_circle + delta_x 
        angle_wall_final = np.degrees(-angle_circle_rebound)
        
        return x_circle, y_circle, x_wall_final, angle_wall_final
```

```python
def reflect_lines(incoming_v, mirror_v, verbose=False):
    EPS = 1e-10
    incoming = incoming_v/np.linalg.norm(incoming_v)
    mirror = mirror_v/np.linalg.norm(mirror_v)

    constant = -np.dot(mirror, incoming)
#     print(f'constant {constant}')
    a = 1
    b = -2*constant*mirror[0]
    c = constant**2 - mirror[1]**2
    x_out = np.roots((a,b,c))
    x_out = [x for x in x_out if not math.isclose(x, -incoming[0])]
#     print(f'xo {x_out}')
    x_out = x_out[0]
    y_out = (constant - x_out*mirror[0])/mirror[1]
    outgoing = (x_out, y_out)
    if verbose:
        print(f'incoming {incoming}')
        print(f'mirror {mirror}')
        print(f'outgoing {outgoing}')
        print('---')
        print(f'dot incoming {np.dot(mirror, incoming)}')
        print(f'dot outgoing {np.dot(mirror, outgoing)}')
        print('---')                          
    return x_out, y_out
```

```python
def get_circle_contact(x_wall_start, angle=None, slope=None):
    """Find the point at which a ball leaving wall will contact the circle.
    Note that this function assumes circle centered at (0,2) with radius 1."""
    if slope is None:
        angle_init = np.radians(angle)
        slope = 1/np.tan(angle_init)
#         print(f'slope {slope}')
    a = slope**2 + 1
    b = 2*slope*x_wall_start - 4
    c = x_wall_start**2 + 3    
    try:
        y_circle = np.roots((a,b,c))
    except:
        d=0.5/np.abs(slope)
        return x_wall_start+d*slope, d, x_wall_start+d*slope, d, angle
    y_circle = [y for y in y_circle if not isinstance(y, complex)]
    if len(y_circle)==0:
        d=0.5/np.abs(slope)
        return x_wall_start+d*slope, d, x_wall_start+d*slope, d, angle
    else:
        y_circle = min(y_circle)
        x_circle = x_wall_start + slope*y_circle

#         print(f'circle {x_circle, y_circle}')
        radial_vector = (x_circle, y_circle-2) 
        radial_vector = radial_vector/np.linalg.norm(radial_vector)
        
        incoming_vector = (x_circle - x_wall_start, y_circle)
        incoming_vector = incoming_vector/np.linalg.norm(incoming_vector)
#         # compare dot products
#         constant = -np.dot(radial_vector, incoming_vector)
#         a=1
#         b=-2*constant*radial_vector[0]
#         c=constant**2 - radial_vector[1]
#         x_out = np.roots((a,b,c))
#         y_out = np.sqrt(1-x_out**2)
#         print(f'xo {x_out}')
#         print(f'yo {y_out}')
#         angle_radius_horizon = np.arccos(radial_vector[0])
#         angle_triangle = 2*np.pi - angle_init - angle_radius_horizon
#         angle_circle_rebound = angle_init - (2*np.pi - angle_triangle)


        x_out, y_out = reflect_lines(incoming_vector, radial_vector)
        angle_circle_rebound = np.arctan(y_out/x_out)
        if y_out>=0:
#             print('leaving from circle!')
            d = 0.5/np.abs(x_out)
            return x_circle, y_circle, x_circle + d*x_out, y_circle + d*y_out, np.degrees(angle_circle_rebound)
#         print(f'off circle {x_out, y_out}')
#         print(f'angle_circle_rebound {np.degrees(angle_circle_rebound)}')
        delta_x = -y_circle/y_out*x_out
        x_wall_final = x_circle + delta_x 
        angle_wall_rads = np.arctan(-y_out/x_out)
        angle_wall_final = np.degrees(angle_wall_rads) 
        return x_circle, y_circle, x_wall_final, 0, angle_wall_final
```

```python
def kickoff(x_wall, x_start=INITIAL_POSITION[0], y_start=INITIAL_POSITION[1]):
    angle = np.degrees(np.arctan(y_start/(x_wall - x_start)))
    state = (x_start, y_start, x_wall, 0, angle)
    positions = []
    positions += [(state[0], state[1]), (state[2], 0)]
    while math.isclose(state[3], 0.0):    
        state = get_circle_contact(state[2], angle=state[-1])
        positions += [(state[0], state[1]), (state[2], state[3])]
        
    df_pos = pd.DataFrame(positions, columns=['x','y'])
    df_pos['radial_dist'] = np.sqrt((df_pos['x'] - CIRCLE_CENTER[0])**2 + (df_pos['y'] - CIRCLE_CENTER[1])**2)
    df_pos['label'] = 'trajectory'    
    
    df_circle = pd.DataFrame({'x' : np.arange(-0.99,0.99,0.01)})
    df_circle['y'] = 2 - np.sqrt(1 - df_circle['x']**2)
    df_circle['label'] = 'circle'    
    df_pos = pd.concat([df_pos, df_circle])
    return df_pos
```

```python
df_pos = kickoff(-0.9)
pn.ggplot(df_pos, pn.aes(x='x', y='y', color='label')) + pn.geom_path() + pn.coord_fixed(ratio = 1)
```
