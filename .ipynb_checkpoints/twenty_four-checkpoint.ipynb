```python
def memoize(f):
    memo = {}
    def helper():
        if (x,y) not in memo:
#             print(f'memo {x,y}')
            memo[(x,y)] = f(x,y)
        return memo[(x,y)]
    return helper    
    
# @memoize
def combine_nums(numlist, target, operations=(), solutions=()):
    if len(numlist)==1:
        if numlist[0]==target:
            if operations[0] not in solutions:
#                 print(f'current: {solutions}')
                print(f'solution: {operations}')
#                 print('---')
                solutions += operations
#             else:
#                 print(f'seen {operations}')
    else:
        for pair in itertools.combinations(range(len(numlist)), 2):
            pv = [numlist[i] for i in pair]
            remainder = [n for i,n in enumerate(numlist) if i not in pair]
            
            new_list_add = [sum(pv)] + remainder
            solutions = combine_nums(new_list_add, target,
                                     operations=operations+(f'{pv[0]}+{pv[1]}',),
                                     solutions=solutions,
                                    )

            new_list_mult = [pv[0]*pv[1]] + remainder
            solutions = combine_nums(new_list_mult, target,
                                     operations=operations+(f'{pv[0]}*{pv[1]}',),
                                     solutions=solutions,
                                    )

            new_list_sub0 = [pv[0] - pv[1]] + remainder
            solutions = combine_nums(new_list_sub0, target,
                                     operations=operations+(f'{pv[0]}-{pv[1]}',),
                                     solutions=solutions,
                                    )

            if pv[1] != pv[0]:
                new_list_sub1 = [pv[1] - pv[0]] + remainder
                solutions = combine_nums(new_list_sub1, target,
                                         operations=operations+(f'{pv[1]}-{pv[0]}',),
                                         solutions=solutions,
                                        )

            if pv[1] != 0:
                new_list_div0 = [pv[0] / pv[1]] + remainder
                solutions = combine_nums(new_list_div0, target,
                                     operations=operations+(f'{pv[0]}/{pv[1]}',),
                                     solutions=solutions,
                                    )

            if pv[0] != 0 and pv[1] != pv[0]:
                new_list_div1 = [pv[1] / pv[0]] + remainder
                solutions = combine_nums(new_list_div1, target,
                                     operations=operations+(f'{pv[1]}/{pv[0]}',),
                                     solutions=solutions,
                                    )

            if not (pv[0]==0 and pv[1]<0):
                new_list_pow0 = [pv[0] ** pv[1]] + remainder
                solutions = combine_nums(new_list_pow0, target,
                                         operations=operations+(f'{pv[0]}^{pv[1]}',),
                                     solutions=solutions,
                                    )

            if not (pv[1]==0 and pv[0]<0) and pv[1] != pv[0]:
                new_list_pow1 = [pv[1] ** pv[0]] + remainder
                solutions = combine_nums(new_list_pow1, target,
                                         operations=operations+(f'{pv[1]}^{pv[0]}',),
                                         solutions=solutions,
                                        )
    return solutions
```

```python
def memoize(f):
    memo = {}
    def helper(x,y):
        if (x,y) not in memo:
            memo[(x,y)] = f(x,y)
        return memo[(x,y)]
    return helper    
    
@memoize
def combine_nums(numlist, target):
    if len(numlist)==1:
        if numlist[0]==target:
            return (('',),)
        else:
            return (False,)
    else:
        solutions=()
        for pair in itertools.combinations(range(len(numlist)), 2):
            pv = tuple(numlist[i] for i in pair)
            remainder = tuple(n for i,n in enumerate(numlist) if i not in pair)
            
            new_list = (sum(pv),) + remainder
            b = combine_nums(new_list, target)
            solutions += tuple((f'{pv[0]}+{pv[1]}',) + x for x in b if x is not False)

            new_list = (pv[0]*pv[1],) + remainder
            b = combine_nums(new_list, target)
            solutions += tuple((f'{pv[0]}*{pv[1]}',) + x for x in b if x is not False)

            new_list = (pv[0] - pv[1],) + remainder
            b = combine_nums(new_list, target)
            solutions += tuple((f'{pv[0]}-{pv[1]}',) + x for x in b if x is not False) 

            if pv[1] != pv[0]:
                new_list = (pv[1] - pv[0],) + remainder
                b = combine_nums(new_list, target)
                solutions += tuple((f'{pv[1]}-{pv[0]}',) + x for x in b if x is not False) 

            if pv[1] != 0:
                new_list = (pv[0] / pv[1],) + remainder
                b = combine_nums(new_list, target)
                solutions += tuple((f'{pv[0]}/{pv[1]}',) + x for x in b if x is not False) 

            if pv[0] != 0 and pv[1] != pv[0]:
                new_list = (pv[1] / pv[0],) + remainder
                b = combine_nums(new_list, target)
                solutions += tuple((f'{pv[1]}/{pv[0]}',) + x for x in b if x is not False) 

            if not (pv[0]==0 and pv[1]<0) and pv[1]<10:
                power = pv[0] ** pv[1]
                if not isinstance(power, complex):                
                    new_list = (pv[0] ** pv[1],) + remainder
                    b = combine_nums(new_list, target)
                    solutions += tuple((f'{pv[0]}^{pv[1]}',) + x for x in b if x is not False) 

            if not (pv[1]==0 and pv[0]<0) and pv[0]<10:
                power = pv[1] ** pv[0]
                if not isinstance(power, complex):
                    new_list = (pv[1] ** pv[0],) + remainder
                    b = combine_nums(new_list, target)
                    solutions += tuple((f'{pv[1]}^{pv[0]}',) + x for x in b if x is not False) 
    return tuple(set(solutions))
```
